```python
import math
```

```python
def sign(x):
    return math.copysign(1,x)
```

```python
def bisezione(fname,a,b,tol):
    
    fa=fname(a)
    fb=fname(b)
    
    if sign(fa)*sign(fb)>=0:
        print("Non è possibile applicare il metodo di bisezione\n")
        return None,None,None
    
    maxit=math.ceil(math.log2(b-a)/tol)-1
    it=0
    v_xk=[]
    while (b-a)>tol:
        xk=a+(b-a)/2
        v_xk.append(xk)
        it=it+1
        
        fxk=fname(xk)
        
        if fxk==0:
            return xk,it,v_xk
        
        if sign(fa)*sign(fxk)>0:  #si continua a lavorare su [xk,b]
            a=xk
            fa=fxk
        elif sign(fxk)*sign(fb)>0:  #si continua a lavorare su [a,xk]
            b=xk
            fb=fxk
    return xk,it,v_xk
```

```python
import numpy as np
import matplotlib.pyplot as plt

f= lambda x: np.exp(-x)-(x+1)
a=-1.0
b=2.0
tol=1e-14
xx=np.linspace(a,b,200)
plt.plot(xx,f(xx))
plt.show()
zero_b,it_bis,v_xk=bisezione(f,a,b,tol)
print("zero_b",zero_b,"\nnit_b",it_bis)
alfa=0
err_bis=np.abs(np.array(v_xk)-alfa)
print('errore_bisezione',err_bis)
plt.plot(range(it_bis),err_bis,'-o')
```
