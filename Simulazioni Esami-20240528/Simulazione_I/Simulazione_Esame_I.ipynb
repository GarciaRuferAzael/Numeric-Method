```python
from scipy.io import loadmat
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
dati=loadmat('testI.mat')
A=dati["A"]
A=A.astype(float)
b=dati["b"]
b=b.astype(float)
```

```python
#simmetria
if np.all(A.T==A):
    print("Matrice simmetrica")

#positiva
if np.all(npl.eigvals(A) > 0):
    print("Matrice definita positiva")

#quadrata
if A.shape[0]==A.shape[1]:
    print("Matrice quadrata")

print(A.shape[0],"Numero di colonne")
if A.shape[0] >= 300:
    print("Matrice di grandi dimesioni")
else:
    print("Matrice di piccole dimensioni")

dim=A.shape[0]*A.shape[1]
nnz=np.count_nonzero(A)
print(dim,"Dimensione")
print(nnz,"Numero di non zeri")
nz=np.count_nonzero(A)/(A.shape[0]*A.shape[1])
print("Numero elementi diversi da zero =",nz)
if np.count_nonzero(A) / nz > 0.33:
    print("Matrice sparsa")
plt.spy(A)


```
